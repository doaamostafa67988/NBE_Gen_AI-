```python
import csv
from selenium import webdriver
import time
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# Initialize the WebDriver (Firefox in this case)
driver = webdriver.Firefox()

# Step 1: Navigate to Facebook and login
url = 'https://www.facebook.com'
driver.get(url)
time.sleep(3)

# Enter your Facebook login credentials
email = driver.find_element(By.NAME, "email")
pswrd = driver.find_element(By.NAME, "pass")
email.send_keys("doaaali2979@gmail.com")
time.sleep(1)
pswrd.send_keys("small56et")
time.sleep(3)
pswrd.submit()  # Submit the form

time.sleep(10)

# Step 2: Go to the specific post URL
post_url = 'https://www.facebook.com/NBE1898'
driver.get(post_url)

# Wait for the post to load
WebDriverWait(driver, 100).until(
    EC.presence_of_element_located((By.XPATH,"//div[contains(@aria-label, 'Comment by ')]//div[contains(@class, 'xdj266r x11i5rnm xat24cr x1mh8g0r x1vvkbs')]"))
)

# Step 3: Retrieve comments from the post
try:
    # Prepare a set to store unique comment data (to avoid duplicates)
    unique_comments = set()

    # Scroll down to load more comments if present
    last_height = driver.execute_script("return document.body.scrollHeight")  # Get the initial height
    scroll_delay = 5  # Delay between scrolls
    scroll_count = 20  # Number of scrolls to perform

    for _ in range(scroll_count):
        # Find all comment elements within the post
        comments = driver.find_elements(By.XPATH, "//div[contains(@aria-label, 'Comment by ')]//div[contains(@class, 'xdj266r x11i5rnm xat24cr x1mh8g0r x1vvkbs')]")
        
        # Get all the text of the comments and add them to the set
        all_comments_text = [comment.text for comment in comments]
        for text in all_comments_text:
            if text:  # Avoid adding empty comments
                unique_comments.add(text)

        # Scroll to the bottom of the page
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        
        # Wait for the new content to load
        time.sleep(scroll_delay)

        # Check if new content was loaded
        new_height = driver.execute_script("return document.body.scrollHeight")
        if new_height == last_height:
            print("No more content to load.")
            break
        last_height = new_height
    
    # Step 4: Save the extracted comments to a CSV file
    file_path = 'C:/Users/HP/Downloads/comments.csv'  # Change the path if needed
    with open(file_path, mode='w', newline='', encoding='utf-8') as file:
        writer = csv.writer(file)
        writer.writerow(["Comment"])  # Write header
        for comment in unique_comments:
            writer.writerow([comment])
    
    print(f"Comments extracted and saved to: {file_path}")

except Exception as e:
    print(f"An error occurred: {e}")

#finally:
    # Step 5: Close the WebDriver
  #  driver.quit()
```

```python
import csv
from selenium import webdriver
import time
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# Initialize the WebDriver (Firefox in this case)
driver = webdriver.Firefox()

# Step 1: Navigate to Facebook and login
url = 'https://www.facebook.com'
driver.get(url)
time.sleep(3)

# Enter your Facebook login credentials
email = driver.find_element(By.NAME, "email")
pswrd = driver.find_element(By.NAME, "pass")
email.send_keys("doaaali2979@gmail.com")
time.sleep(1)
pswrd.send_keys("small56et")
time.sleep(3)
pswrd.submit()  # Submit the form

# Wait for the page to load after login
time.sleep(10)

# Step 2: Go to the specific post URL
post_url = 'https://www.facebook.com/NBE1898/posts/pfbid0k63TEkzoJxbTdfAPEWtcJkp7C1g3LFynnEdVKbpbmNPnqqA5Jne5jRGpQF3rTY9Kl'
driver.get(post_url)

# Wait for the page to load
time.sleep(5)

# Step 3: Retrieve comments from the post
try:
    # Wait until comments are loaded
    WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.XPATH, "//div[contains(@aria-label, 'Comment by')]"))
    )
    
    # Find all comment elements
    comments = driver.find_elements(By.XPATH, "//div[contains(@aria-label, 'Comment by')]")
    
    # Prepare a list to store the comment data
    data = [["Comment"]]
    
    # Scroll down to load more comments if present
    last_height = driver.execute_script("return document.body.scrollHeight")
    
    while True:
        # Loop through the comments and check for 'See more' buttons
        for comment in comments:
            try:
                # Look for the "See more" button in the comment
                see_more_button = comment.find_element(By.XPATH, ".//div[text()='See more']")
                
                # Click the "See more" button if found
                if see_more_button:
                    driver.execute_script("arguments[0].click();", see_more_button)
                    time.sleep(2)  # Wait for the expanded comment to load
            except Exception as e:
                # Ignore if there's no "See more" button
                pass
        
        # Get all the text of the comments after clicking "See more"
        all_comments_text = [comment.text for comment in comments]
        
        # Add each comment to the data list
        for text in all_comments_text:
            data.append([text])
        
        # Scroll down to load more comments
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(3)  # Wait for new comments to load
        new_height = driver.execute_script("return document.body.scrollHeight")
        
        # Break the loop if no more comments are loaded
        if new_height == last_height:
            break
        
        last_height = new_height
    
    # Step 4: Save the extracted comments to a CSV file
    file_path = '/content/extracted_comments.csv'  # Change the path if needed
    with open(file_path, mode='w', newline='', encoding='utf-8') as file:
        writer = csv.writer(file)
        writer.writerows(data)
    
    print(f"Comments extracted and saved to: {file_path}")

except Exception as e:
    print(f"An error occurred: {e}")

# Step 5: Close the WebDriver
#driver.quit()
```
